```python
import sys
reload(sys)
sys.setdefaultencoding("utf-8")

import csv
import numpy as np
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

df = pd.read_csv('train.csv')
```

```python

from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
y = df['Survived']
```

```python
X = df.drop('Survived', axis=1)
```

```python
X
```

# Here preprocess yor data

```python
sexes = X['Sex'].as_matrix()
l = LabelEncoder()
l.fit(sexes)
sexes_tr = l.transform(sexes)

pclasses = X['Pclass'].as_matrix()
```

```python
numpy_data = np.column_stack((sexes_tr, pclasses))
```

```python
X_train, X_test, y_train, y_test = train_test_split(numpy_data, y, random_state=42)
```

# Here choose classifier(s) and parameters

```python
cls = KNeighborsClassifier()
cls.fit(X_train, y_train)
```

# And check the score

```python
pred = cls.predict(X_test)
accuracy_score(y_test, pred)
```
